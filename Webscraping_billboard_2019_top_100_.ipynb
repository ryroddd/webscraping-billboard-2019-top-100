# Web Scraping to SQL Assignment

### I first imported the libraries that I needed for the Assignment

```python
import pandas as pd
import lxml.html
import requests
import sqlite3
```

#### I use the requests library to save the url to the r variable

```python
r = requests.get("https://www.billboard.com/charts/year-end/2019/top-artists")
```

#### The html is used to detect the html syntax

```python
html = lxml.html.fromstring(r.text)
```

#### I figured out the pathway to the ranking values and display it as a text

```python
html.xpath('//div[@class="ye-chart-item__rank"]')[0].text
```

#### Here I gather all the value that is in the ranking column

```python
for location_cell in html.xpath('//div[@class="ye-chart-item__rank"]'):
    print(location_cell, location_cell.xpath('string()'))
```

#### I found the artist value or name

```python
html.xpath('//div[@class="ye-chart-item__title"]')[0].getchildren()[0].text
```

#### Here I gather all the value that is in the artist column

```python
for location_cell in html.xpath('//div[@class="ye-chart-item__title"]'):
    print(location_cell, location_cell.xpath('string()'))
```

#### I noticed the new line character is present along with the value. I have to remove them before I continue on in my assignment.

```python
[x.xpath("string()") for x in html.xpath('//div[@class="ye-chart-item__rank"]')]
```

#### I used the map() and strip() to remove the new line characters in the value

```python
rank = [x.xpath("string()") for x in html.xpath('//div[@class="ye-chart-item__rank"]')]
official_rank = map(str.strip, rank)
```

#### Again, the new line character is present along with the value. I have to remove them.

```python
[x.xpath("string()") for x in html.xpath('//div[@class="ye-chart-item__title"]')]
```

#### Like last time, I used map() and strip() to remove the new line character

```python
name = [x.xpath("string()") for x in html.xpath('//div[@class="ye-chart-item__title"]')]
official_name = map(str.strip, name)
```

#### Once I clean up the data, I then put them in a data frame.

```python
df = pd.DataFrame({
    "Rank"  :official_rank,
    "Artist":official_name
})
df
```

#### I Connect to the SQLite database and insert the top 5 artist

```python
db = sqlite3.connect(":memory:")
cursor = db.cursor()
cursor.execute("""
    CREATE TABLE LIST(RANK, ARTIST)
""")

for row in df[df.Artist.isin(["Post Malone"])].itertuples():
    insert_sql_syntax = """
        INSERT INTO LIST(
        RANK,
        ARTIST)
        VALUES (?,?)
    """
    cursor.execute(insert_sql_syntax, row[1:])
    
for row in df[df.Artist.isin(["Ariana Grande"])].itertuples():
    insert_sql_syntax = """
        INSERT INTO LIST(
        RANK,
        ARTIST)
        VALUES (?,?)
    """
    cursor.execute(insert_sql_syntax, row[1:])
    
for row in df[df.Artist.isin(["Billie Eilish"])].itertuples():
    insert_sql_syntax = """
        INSERT INTO LIST(
        RANK,
        ARTIST)
        VALUES (?,?)
    """
    cursor.execute(insert_sql_syntax, row[1:])
    
for row in df[df.Artist.isin(["Khalid"])].itertuples():
    insert_sql_syntax = """
        INSERT INTO LIST(
        RANK,
        ARTIST)
        VALUES (?,?)
    """
    cursor.execute(insert_sql_syntax, row[1:])
    
for row in df[df.Artist.isin(["Drake"])].itertuples():
    insert_sql_syntax = """
        INSERT INTO LIST(
        RANK,
        ARTIST)
        VALUES (?,?)
    """
    cursor.execute(insert_sql_syntax, row[1:])
db.commit()
#db.close()
```
